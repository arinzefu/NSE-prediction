```python
# Define the split points
split_date1 = '2021-01-01'
split_date2 = '2022-07-01'

# Split the data into training, validation, and test sets
train_data = df[df['Date'] < split_date1]
val_data = df[(df['Date'] >= split_date1) & (df['Date'] < split_date2)]
test_data = df[df['Date'] >= split_date2]
```

```python
# Plot the entire dataset
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Price'], label='Full Dataset', color='blue')

# Plot the training, validation, and test sets
plt.scatter(train_data['Date'], train_data['Price'], label='Training Set', color='green', marker='o')
plt.scatter(val_data['Date'], val_data['Price'], label='Validation Set', color='orange', marker='o')
plt.scatter(test_data['Date'], test_data['Price'], label='Test Set', color='red', marker='o')

# Add labels and legend
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Dataset Split: Training, Validation, and Test Sets')
plt.legend()

# Show the plot
plt.show()
```

```python
# Set 'Date' as the index column
train_data.set_index('Date', inplace=True)
val_data.set_index('Date', inplace=True)
test_data.set_index('Date', inplace=True)

# Split data into features and target variables
X_train = train_data.drop(columns=['Price'])
y_train = train_data['Price']

X_val = val_data.drop(columns=['Price'])
y_val = val_data['Price']

X_test = test_data.drop(columns=['Price'])
y_test = test_data['Price']
```

```python

from sklearn.preprocessing import StandardScaler


# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler on the training data and transform the training, validation, and test data
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python
sequence_length = 10

def create_sequences_with_time(X, y, sequence_length):
    X_seq, y_seq, timestamps = [], [], []
    for i in range(len(X) - sequence_length + 1):
        X_seq.append(X[i:i+sequence_length])
        y_seq.append(y[i+sequence_length-1])
        timestamps.append(X.index[i+sequence_length-1].strftime('%Y-%m-%d'))
    return np.array(X_seq), np.array(y_seq), np.array(timestamps)

X_train_seq, y_train_seq, train_timestamps = create_sequences_with_time(X_train_scaled, y_train.values, sequence_length)
X_val_seq, y_val_seq, val_timestamps = create_sequences_with_time(X_val_scaled, y_val.values, sequence_length)
```
